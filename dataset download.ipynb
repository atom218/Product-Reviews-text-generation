```python
# Step 2: Create a project folder in your Drive
import os

project_dir = '/content/drive/MyDrive/VisualReview'
os.makedirs(f'{project_dir}/data/raw', exist_ok=True)

print("Folder created at:", project_dir)
```

```python
# Step 4: Load and save the REVIEWS file
from datasets import load_dataset
import json

print("Loading reviews... (this may take a few minutes)")

reviews = load_dataset(
    "McAuley-Lab/Amazon-Reviews-2023",
    "raw_review_Electronics",
    split="full"
)

# Save to Drive as JSONL
output_path = f'{project_dir}/data/raw/Electronics_reviews.jsonl'
with open(output_path, 'w') as f:
    for row in reviews:
        f.write(json.dumps(row) + '\n')

print(f"Reviews saved! Total records: {len(reviews)}")
```

```python
# Step 5 (FIXED): Load METADATA via direct JSON file
from datasets import load_dataset
import json

print("Loading metadata via direct JSONL paths...")

meta = load_dataset(
    "json",
    data_files="https://huggingface.co/datasets/McAuley-Lab/Amazon-Reviews-2023/resolve/main/raw/meta/Electronics.jsonl.gz",
    split="train"
)

output_path = f'{project_dir}/data/raw/Electronics_meta.jsonl'
with open(output_path, 'w') as f:
    # Sampling 10,000 for efficiency
    for row in meta.select(range(min(10000, len(meta)))):
        f.write(json.dumps(row) + '\n')

print(f"Metadata saved! Total records sampled: {min(10000, len(meta))}")
```
